```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('dulieuxettuyendaihoc.csv', 
               header=0, delimiter=',', 
               encoding='utf-8')
df
```

```python
#in ra thông tin
df.info()
```

```python
df=df[['T5','T6','GT','DT','KV','KT','NGONNGU','TOANLOGICPHANTICH','GIAIQUYETVANDE','NGAYTHI','DINHHUONGNGHENGHIEP']]
```

```python
#đổi tên
df.rename(columns={'TOANLOGICPHANTICH':'LOGIC',
                   'GIAIQUYETVANDE':'UNGXU',
                   'DINHHUONGNGHENGHIEP':'HUONGNGHIEP'}, 
                   inplace=True)
df
```

```python
#Xóa bỏ các dòng dữ liệu rỗng
df.dropna(how='all', inplace=True)
df
```

```python
#xóa bỏ các dòng dữ liệu bị trùng
df.drop_duplicates(inplace=True)
df
```

```python
#Dùng heatmap để trực quan hóa dữ liệu thiếu
plt.figure(figsize=(10,6))
sns.heatmap(df.isna().transpose(), cmap='YlGnBu',
    cbar_kws={'label':'Dữ liệu thiếu'})
plt.savefig('missingdata.png', dpi=100)
'''
    Note: với dữ liệu bị thiếu: (1) càn xác định biến số nào bị thiếu
    (2) mức độ thiếu dữ liệu
    (3) có cần phải xư lý không

    => dựa vào biều đồ ta thấy: biến DT bị thiếu
'''
```

```python
#Điền giá trị thiếu
df['DT'].fillna('KINH', inplace=True)
#lưu ý: với biến định tính ta có thể thay bằng giá trị yếu vị (mode)
#df['DT'].fillna(df['DT'].mode()[0], inplace=True)
df
```

```python
#Điền thiếu giá trị phần NGONNGU bằng 0 (nếu có)
df['NGONNGU'].fillna(0, inplace=True)
#Điền thiếu giá trị phần LOGIC bằng trung bình (nếu có)
df['LOGIC'].fillna(df['LOGIC'].mean(), inplace=True)
#Điền thiếu giá trị phần UNGXU bằng trung vị (nếu có)
df['UNGXU'].fillna(df['UNGXU'].median(), inplace=True)
#Lưu ý: với biến định lượng thì ta nên thay bằng trung vị
df
```

```python
'''
Phần 2: Kỹ thuật Feature Engineering
(thường dùng cho Machine Learning)
Nếu chỉ là xửu lý phân tích dữ liệu thì ta gọi là New Attribute

Đây là kỹ thuật tạo thêm hoặc biến đổi số liệu sẵn có thành
các biến số mới phù hợp với nghiệp vụ để phân tích

Tạo biến TBTOAN: trung bình toán lớp 12
'''
df['TBTOAN']=(df['T5']+df['T6'])/2
df
```

```python
#Tạo biến XEPLOAI: đánh giá môn toán dựa trên df['TBToan]
df.loc[df['TBTOAN']<5.0, 'XEPLOAI']='FAIL'

df.loc[(df['TBTOAN']>=5.0) &
       (df['TBTOAN']<7.0), 'XEPLOAI']='FAIR'

df.loc[(df['TBTOAN']>=7.0) &
       (df['TBTOAN']<9.0), 'XEPLOAI']='GOOD'

df.loc[df['TBTOAN']>=9.0, 'XEPLOAI']='EXCEL'
df
```

```python
#tìm dữ liệu theo yêu cầu - tìm loại 2NT trên KV
df[df['KV'] == '2NT']
```

```python
'''
Tạo biến nhóm khối thi: NHOMKT thõa mãn 
A1: G1
C: G3
D1: G3
A: G1
B: G2
'''
dict_map={
    'A1':'G1','C':'G3','D1':'G3','A':'G1','B':'G2'
}
df['NHOMKT']=df['KT'].map(dict_map)
df
```

```python
'''
#Tạo bién số điểm cộng: CONG
Nếu khối thi thuộc nhóm G1, G2 và TBTOAN >= 5.0 thì 1.0 
Ngược lại thì 0.0
'''
def fplus(x,y):
    if (x=='G1' or x=='G2') and (y >= 5.0):
        return 1.0
    else:
        return 0.0

df['CONG']=list(map(fplus,df['NHOMKT'], df['TBTOAN']))
df
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy
import scipy.stats as stats
from statsmodels.stats.weightstats import ztest as ztest
from scipy.stats.stats import pearsonr
from scipy.stats import chi2_contingency
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

```python
df=pd.read_csv('dulieuxettuyendaihoc.csv')
df.head(10)
```

```python
df.rename(columns={'TOANLOGICPHANTICH':'LOGIC',
                   'GIAIQUYETVANDE':'UNGXU',
                   'DINHHUONGNGHENGHIEP':'HUONGNGHIEP'}, 
                   inplace=True)
```

```python
'''
Phân tích suy diễn (ìnerrential statistics)
lý do tạo sai cần suy diễn: kết luạn dựa trên dữ liệu sample (mẫu),
nhưng kết luận thì được hiểu là áp dụng cho tổng thể
    Câu hỏi đặt ra: kết luận trên mẫu đso có phù hợp với tổng thể hay không

Khi phân tích suy diễn cần lưu ý:
    1. Xác định giả thuyết H0
    2. Các giả định hay các điều kiện về dữ liệu và môi trường để áp dụng kiểm định
        - Lưu ý vè giả địnnh phân phối của biến số: Normal, Student, Poison, Chi-square,....

    3. Thiết lập mứuc tin cậy và sai lầm (alpha): 90%-10%, 95%-5% và 99%-1%
    4. Quy tắc suy diễn:
        4.1 Nếu p-value < alpha: bác bỏ giả thuyết H0
        4.2 nếu ........>......: chấp nhật giả thuyết H0
    5.Kết luận của suy diễn chỉ cho chúng ta biết là có đủ dữ liệu để kết luận cho tổng thể hay không (còn gọi là ý nghĩa thống kê)
'''


'''
Phục vụ 3 bài toán chính sau:
    - Khám phá dữ liệu tương tác qua lại định tính - định tính
    - ...................................định tính - định lượng
    - ...................................định lượng - định lượng
'''
```

```python
'''
ONE SAMPLE T-TEST
Mục đích: kiểm tra 
'''
```

```python
#Câu 1: kiểm tra xem trung bình điểm LOGIC của thí sinh khối C có bằng 4 hay không
#kiểm tra xem n<30 hay không? như bên dưới ta đếm được n=14 nên lấy
dfkhoic=df.loc[df['KT']=='C']
dfkhoic['LOGIC']
```

```python
#Với one sample T-test thì giả thuyế H0: muy=4.0
#mặc định mức tin cậy là 95% và mức sai lầm là 5%
stats.ttest_1samp(dfkhoic['LOGIC'], popmean=4.0)
#kết luận: đủ điều kiện để chấp nhạn giả thuyết H0, điểm trung bình của môn thi LOGIC bằng 4, hay
#nói cách khác chấp nhận H0 ở mức sai lầm 5%
```

```python
#sinh viên kiểm tra  xem điểm trung bình của môn thi UNGXU của khối thi C có bằng 5.0 hay không?
#đếm các giá trị nếu dưới 30 thì tính tiếp (n<30)
dfkhoic=df.loc[df['KT']=='C']
dfkhoic['UNGXU']
```

```python
stats.ttest_1samp(dfkhoic['UNGXU'], popmean=5.0)
#=>>> chấp nhận H0
#kết luận: đủ điều kiện để chấp nhạn giả thuyết H0, điểm trung bình của môn thi UNGXU bằng 5, hay
#nói cách khác chấp nhận H0 ở mức sai lầm 5%
```

```python
#Câu 2: hãy kiểm tra xem trung bình điểm thi LOGIC và trung bình điểm thi UNGXU của thí sinh khối C có bằng nhau hay không?
#H0: mean_LOGIC - mean_UNGXU =0

dfkhoic=df.loc[df['KT']=='C']
stats.ttest_ind(dfkhoic['LOGIC'], dfkhoic['UNGXU'], equal_var=True)
```

```python
'''
Kết luận: do alpla = 0.05 và p-value = 0.3111 suy ra đủ dữ liệu để nói rằng triung bình LOGIC bằng trung bình UNGXU
hay nói cách khác là chấp nhận H0 ở mức sai lầm 5%
'''
```

```python
'''
Sinh viên làm tương tự:
kiểm tra xem trung binh UNGXU có bằng trung bình NGONNGU cho thí sinh khối C hay không
'''
dfkhoic=df.loc[df['KT']=='C']
stats.ttest_ind(dfkhoic['UNGXU'], dfkhoic['NGONNGU'], equal_var=True)
#Kết luận: do alpla = 0.05 và p-value = 0.014 suy ra không đủ dữ liệu để nói rằng trung bình LOGIC bằng trung bình UNGXU
#hay nói cách khác là không chấp nhận H0 ở mức sai lầm 5%
```

```python
'''
Hãy kiểm tra xem có sự phụ thuộc nào giữa việc sinh viên có định hướng nghề nghiệp
và giới tính khi thí sinh đăng ký dự thi hay không
'''
crosdata=pd.crosstab(df['GT'], [df['HUONGNGHIEP']], rownames=['GT'], colnames=['HUONGNGHIEP'])
crosdata
```

```python
'''
chi_square TEST
mục đích: kiểm tra sự độc lập của 2 biến định tính
H0: khong có sự khác biệt giữa 2 biến định tính
'''
```

```python
'''
Hãy kiểm tra xem có sự phụ thuộc nào giữa khối thi và khu vực khi đăng ký dự thi hay khoong
'''
crosdata=pd.crosstab(df['KV'], [df['KT']], rownames=['KV'], colnames=['KT'])
crosdata
```

```python
stat, p, dof, expected= chi2_contingency(crosdata)

#interpnet p_value
alpha=0.05
print ("p value is: " + str(p))
if p<=alpha:
    print('Dependent (reject H0)')
else:
    print('Independent (H0 holds true)')
```

```python
''''
Két luận:bác bỏ H0 vì p_value = 0.0201246 < alpha = 0.05
tức là: không đủ dữ liệu để nói rằng KV và KT là độc lập hay có sự quan hệ giữa KT và KV
'''
```
