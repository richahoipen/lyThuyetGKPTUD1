```python
import pandas as pd
import seaborn as sns 
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('orginal_sales_data_edit.csv', encoding = 'unicode_escape', header = 0, delimiter=',')
```

```python
# hiển thị 5 dòng đầu tiên
df.head()
'''   
ORDERNUMBER: Mã hóa đơn
QUANTITYORDERED: Số lượng mua sản phẩm
PRICEEACH : ĐƠN GIÁ
ORDERLINENUMBER: Dòng thông tin mua sản phẩm trên hóa đơn, trên hóa đơn có thể có nhiều dòng thông tin
mua các sản phẩm khác nhau
SALES: Giá trị tiền mua sản phẩm
ORDERDATE: Ngày đặt hàng
STATUS: Trạng thái cho biết tình trạng món hàng
COUNTRY: Chi nhánh tại quốc gia
DEALSIZE: Phân loại kích thước theo giá trị hàng mua
POSTALCODE	: mã bưu điện

'''
```

```python
#tính tổng giá thành
df['TOTALPRICE'] = df['QUANTITYORDERED'] * df['PRICEEACH']
df
```

```python
df.dropna(how='all', inplace=True)
```

```python
df.drop_duplicates(inplace=True)
```

```python
#trực quan xem dữu liệu bị thiếu
#trực quan xem dữ liệu nào không bị thiếu dữu liệu nào cần xử lý
plt.figure(figsize = (10,6))
sns.heatmap(df.isna().transpose(), cmap='YlGnBu', cbar_kws={'label':'Dữ liệu thiếu'})
plt.savefig('missingdata1.png', dpi=100)
```

```python
plt.figure(figsize =(10,6))
sns.displot(
    data=df.isna().melt(value_name="missing"),
    y="variable",
    hue= "missing",
    multiple="fill",
    aspect=1.25
)
plt.savefig("my_missing_value_2.png", dpi=100)
'''     
  các cột màu đậm là cột đang bị thiếu(TRUE)
  
'''
```

```python
#điền gái trị thiếu
df['ADDRESSLINE2'].fillna('unknown', inplace=True)
df['STATE'].fillna('unknown', inplace=True)
df['POSTALCODE'].fillna('0', inplace=True)
df['TERRITORY'].fillna('unknown', inplace=True)
'''
  với giá trị là biến định tính có thể thay thế bằng giá trị yếu vị mode()
    df['DT'].fillna(df['DT'].mode(), inplace=True)
'''
```
