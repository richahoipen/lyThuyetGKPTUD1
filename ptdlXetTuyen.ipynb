```python
#doi ten TOANLOGICPHANTICH thành LOGIC
df.rename(columns={'TOANLOGICPHANTICH':'LOGIC'},inplace=True)
df.loc[10:20]
```

```python
df['DT'].fillna('KINH',inplace=True)
df.head()
```

```python
df['TBTOAN']=(df['T5']+df['T6'])/2
df.head()
```

```python
df[4:20]
```

```python
#tao cột xếp loại dựa vào TBTOAN
df.loc[df['TBTOAN']<5.0,'XEPLOAI']='FAIL'
df.loc[(df['TBTOAN']>=5.0) & (df['TBTOAN']<7.0),'XEPLOAI']='AVERAGE'
df.loc[(df['TBTOAN']>=7.0) & (df['TBTOAN']<9.0),'XEPLOAI']='GOOD'
df.loc[df['TBTOAN']>=9.0,'XEPLOAI']='EXCELENT'
df[5:20]
```

```python
df.drop('XEPLOAI', axis=1)
```

```python
#dùng map để tạo cột NHOMKT dựa trên định tính KT
#tạo định tính mới theo định tính trước đó
dict_map={'A1':'G1','C':'G3','D1':'G4','A':'G1','B':'G2'}
#Nếu KT='A1' thì NHOMKT ='G1', nếu KT='C' thì NHOMKT='G3',...
df['NHOMKT']=df['KT'].map(dict_map)
df[1:30]
```

```python
#tạo cột DIEMCONG thoả mãn
#nếu NHOMKT thuộc G1,G2 và TBTOAN>=5.0 thì 1.0, ngược lại là 0.0
def diemcong(row):
    if row['NHOMKT'] in ['G1', 'G2'] and row['TBTOAN'] >= 5.0:
        return 1.0
    else:
        return 0.0

df['DIEMCONG'] = df.apply(diemcong, axis=1)
df[5:20]
```
